```python
#  Student Performance Dataset Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1 Load the dataset
df = pd.read_csv("Students Performance Dataset.csv")

# 2 Basic dataset info
print("Dataset Shape:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nMissing Values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)

# 3 Descriptive statistics
print("\nNumerical Summary:\n", df.describe())
print("\nCategorical Summary:\n", df.describe(include='object'))

# 4 Correlation analysis
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 5 Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Gender", palette="Set2")
plt.title("Gender Distribution")
plt.show()

# 6 Department distribution
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="Department", palette="Set3")
plt.title("Department Distribution")
plt.xticks(rotation=45)
plt.show()

# 7 Grade distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Grade", palette="Set1", order=sorted(df['Grade'].unique()))
plt.title("Grade Distribution")
plt.show()

# 8 Relationship between Study Hours and Total Score
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Study_Hours_per_Week", y="Total_Score", hue="Gender", alpha=0.7)
plt.title("Study Hours vs Total Score")
plt.show()

# 9 Attendance vs Total Score
plt.figure(figsize=(8, 5))
sns.regplot(data=df, x="Attendance (%)", y="Total_Score", scatter_kws={'alpha':0.6})
plt.title("Attendance vs Total Score")
plt.show()

#  Stress Level vs Sleep Hours
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Stress_Level (1-10)", y="Sleep_Hours_per_Night", hue="Gender")
plt.title("Stress Level vs Sleep Hours")
plt.show()

# 11 Average score per department
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Department", y="Total_Score", estimator=np.mean, palette="viridis")
plt.title("Average Total Score per Department")
plt.xticks(rotation=45)
plt.show()

# 12 Grade vs Study Hours boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Grade", y="Study_Hours_per_Week", palette="cool")
plt.title("Study Hours Distribution by Grade")
plt.show()

#  13 Print correlations with Total Score
print("\nCorrelation of features with Total Score:\n")
print(corr["Total_Score"].sort_values(ascending=False))
```

```python
#  Student Gender Distribution by Department
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1 Load dataset
df = pd.read_csv("Students Performance Dataset.csv")

# 2 Total number of students
total_students = len(df)
print(" Total number of students:", total_students)

# 3 Department-wise gender distribution
dept_gender_counts = df.groupby(["Department", "Gender"]).size().reset_index(name="Count")
print("\n Department-wise Male/Female Distribution:\n")
print(dept_gender_counts.pivot(index="Department", columns="Gender", values="Count").fillna(0))

# 4 Plot chart
plt.figure(figsize=(10, 6))
sns.barplot(data=dept_gender_counts, x="Department", y="Count", hue="Gender", palette="Set2")

# Add titles and labels
plt.title("Department-wise Male/Female Student Distribution", fontsize=14)
plt.xlabel("Department", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.xticks(rotation=30)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

```python
#  Five-Point Summary for Students Performance Dataset
import pandas as pd

# 1 Load dataset
df = pd.read_csv("Students Performance Dataset.csv")

# 2 Select only numeric columns
numeric_df = df.select_dtypes(include=["number"])

# 3 Compute five-number summary
five_point_summary = numeric_df.describe(percentiles=[0.25, 0.5, 0.75]).loc[["min", "25%", "50%", "75%", "max"]]

# 4 Display neatly
print(" Five-Point Summary (Min, Q1, Median, Q3, Max):\n")
print(five_point_summary)
```
